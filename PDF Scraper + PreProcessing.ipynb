```python
# Empty list to store series prior to inserting in DataFrame
comb_list = []

# First for loop iterates through every other value in each column and combines them as one
for header in column_headers:
    series_list = [data[0].loc[i, header] for i in range(0, len(data[0]), 2)]
    new_series = pd.Series(series_list, name=header)
    comb_list.append(new_series)

# Second for loop iterates through the every other value from the second row for specific columns
for header in column_headers_2:
    series_list = [data[0].loc[i, header] for i in range(1, len(data[0]), 2)]
    new_series = pd.Series(series_list, name=header)
    comb_list.append(new_series)
    
    
new_df = pd.concat(comb_list, axis=1)
                   
```
