# Single Function to Process Imported PDF Table Data

```python
# Custom function to convert mm:ss.SSS to Timedelta

def custom_to_timedelta(duration_str):
    
    if pd.isna(duration_str):
        return pd.NaT
    elif duration_str in ['DNF', 'DSQ', 'DNS']:
        minutes = 0
        seconds = 0
    else:
        minutes, seconds = map(float, duration_str.split(':'))
        
        
        return pd.to_timedelta(minutes * 60 + seconds, unit='s')
```

```python
def table_sort(data):
    '''
    Race time data is stored in PDFs, tabula can be used to extract the data but the format it is is isn't workable. This
    function re-arranges and pre-processes the data so that it can be then added to a larger data base that is clean and
    consistent for further analysis.
    
    Parameters:
    Takes a DataFrame of an extracted table from PDF containing the race time data from Downhill World Cups. These DataFrames
    all need to be in a particular format for this function to work
    
    Returns:
    The re-arranged and processed dataframe
    '''
    
    column_header = data.columns.tolist()
    
    
    # Checking that 'Rank' and Nr Name / UCI MTB Team are in seperate columns. If not, splits the Rank and re-arranges columns 
    # in format that the main function will be able to process correctly
    if column_header[0] != 'Rank':
        
        data[column_header[0]] = data[column_header[0]].astype(str)
        
        
        # Standard expression that look for numbers up until a period and then takes the rest of the string, splitting these 
        # into two seperate columns
        pattern = r'^(\d+\s*)?(.*)$'
        data[['Rank', 'Nr Name / UCI MTB Team']] = data[column_header[0]].str.extract(pattern)
        
        
        # Delete the original combined column 'Rank Nr Name / UCI MTB Team' and 'Unnamed: 0' column
        data = data.drop([column_header[0], column_header[1]], axis=1)
        
        
        # Re-order the columns into the same order as the they are in the PDF
        column_order_orig = ['Rank', 'Nr Name / UCI MTB Team', 'UCI ID', 'NAT', 'YOB', 'Speed', 'I1 / I2', 'I3 / I4', 'Time', 'Gap']
        data = data[column_order_orig]
        
        
        # Update column_header to reflect the new order of the columns
        column_header = data.columns.tolist()
    
    
    # Empty list to store series prior to inserting in DataFrame
    comb_list = [] 

    
    # First for loop iterates through every other value in each column and combines them as one in a list
    for header in column_header:
        series_list = [data.loc[i, header] for i in range(0, len(data), 2)]
        new_series = pd.Series(series_list, name=header)
        comb_list.append(new_series)
    
    
    # Second for loop iterates through the every other value from the second row for specific columns in a list
    column_header_2 = ['Nr Name / UCI MTB Team', 'I1 / I2', 'I3 / I4']

    for header in column_header_2:
        series_list = [data.loc[i, header] for i in range(1, len(data), 2)]
        new_series = pd.Series(series_list, name=header)
        comb_list.append(new_series)
    
    
    # Stores list in new data frame
    new_df = pd.concat(comb_list, axis=1)
    
    
    # Updated header names for columns and re-naming DataFrame
    new_headers = ['Rank', 'Nr Name', 'UCI ID', 'Nat', 'YOB', 'Speed', 'I1', 'I3', 'Time', 'Gap', 'UCI MTB Team', 'I2', 'I4']
    new_df.columns = new_headers
    
    
    # Standard expression that splits 'Nr Name' string into two seperate columns, the number that comes first - 'World Ranking'
    # and the riders 'Name'
    pattern_name = r'(\d+) (\w+ \w+)'
    new_df[['World Ranking', 'Name']] = new_df['Nr Name'].str.extract(pattern_name)
    
    
    # Standard expression that handles seperation of the 'Split Pos' and 'Split (No.)' storing seperately
    pattern_split_times = r'(\d+:\d+\.\d+) \((\d+)\)$'
    new_df[['Split 1', 'Split 1 Pos']] = new_df['I1'].str.extract(pattern_split_times)
    new_df[['Split 2', 'Split 2 Pos']] = new_df['I2'].str.extract(pattern_split_times)
    new_df[['Split 3', 'Split 3 Pos']] = new_df['I3'].str.extract(pattern_split_times)
    new_df[['Split 4', 'Split 4 Pos']] = new_df['I4'].str.extract(pattern_split_times)
    
    
    # Drop columns that have been seperated into two seperate columns
    new_df = new_df.drop(['Nr Name', 'I1', 'I2', 'I3', 'I4'], axis=1)
    
    
    # With the data now seperated and all on one single row, the column order is specified and the DataFrame re-ordered
    column_order = ['Rank', 'World Ranking', 'Name', 'UCI MTB Team', 'UCI ID', 'Nat', 'YOB', 'Speed', 'Split 1', 'Split 1 Pos', 'Split 2', 'Split 2 Pos', 'Split 3', 'Split 3 Pos', 'Split 4', 'Split 4 Pos', 'Time', 'Gap']
    new_df = new_df[column_order]
        
    
    # Set each column to the most appropriate data type

    new_df['Rank'] = pd.to_numeric(new_df['Rank'], errors='coerce', downcast='integer')

    new_df['World Ranking'] = pd.to_numeric(new_df['World Ranking'], errors='coerce', downcast='integer')

    new_df['UCI ID'] = new_df['UCI ID'].astype(str)

    new_df['YOB'] = pd.to_datetime(proc_df['YOB'], format='%Y')

    
    split_pos = ['Split 1 Pos', 'Split 2 Pos', 'Split 3 Pos', 'Split 4 Pos']
    for split in split_pos:
        new_df[split] = pd.to_numeric(new_df[split], errors='coerce', downcast='float')
    
    #Converts all times into Timedelta format
    split_no = ['Split 1', 'Split 2', 'Split 3', 'Split 4']
    for split in split_no:
        new_df[split] = new_df[split].apply(custom_to_timedelta)

    new_df['Time'] = new_df['Time'].apply(custom_to_timedelta)
    
    return new_df
```
