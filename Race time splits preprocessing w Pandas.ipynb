```python
import pandas as pd
import numpy as np
import chardet
```

# Preprocessing

```python
# Check the encoding of the CSV file 

with open('2022 Lourdes Mens Elite Times.csv', 'rb') as file:
    result = chardet.detect(file.read())
    
print(result['encoding'])
```

```python
# Import the csv file using pandas read_csv() method

race_times_raw_data = pd.read_csv('2022 Lourdes Mens Elite Times.csv', encoding='windows-1252')
race_times_raw_data
```

```python
# View the data types of each column

race_times_raw_data.dtypes
```

```python
# Create a new variable that replicates the original data set, leaving the original import un processed for later reference

race_times = race_times_raw_data.copy()
```

```python
race_times.describe(include='all')
```

```python
# Check the names of columns are correctly stored

race_times.columns.array
```

```python
# Remove blank spaces from titles that have them

race_times.rename(columns=lambda x: x.strip().lower().replace(' ','_'), inplace=True)
```

```python
# Check titles are correct

race_times.columns.array
```

```python
# Check for unique values in 'position' - this column shows the racers position in the race and is unique for each racer

race_times['position'].unique()
```

```python
# Change 'position' data type to a string variable instead on an int

race_times['position'] = race_times['position'].astype(str)
```

```python
# Check for unqique values in 'name'

race_times['name'].unique()

# The names here are in a confusing format, standarise them and create seperate columns for first and last name
# Unsure if this is useful to keep them seperate at this stage
```

```python
# Split strings in column 'name' and store first and last name in seperate columns

race_times[['last_name', 'first_name']] = race_times['name'].str.split(' ', 1, expand=True)

race_times
```

```python
# Standarise the string format

race_times['last_name'] = race_times['last_name'].str.title()
```

```python
race_times = race_times.drop(['name'], axis=1)
race_times
```

```python
race_times.dtypes
```

```python
race_times.describe(include='all')
```

```python
# Examine the unqiue values in 'team' and check for NA values

race_times['team'].unique()

# Blank values are stored as nan
```

```python
# Replace nan values with pandas recognised format

race_times['team'] = np.where(race_times['team']=='nan', pd.NA, race_times['team'])

race_times['team'].unique()

# np.where() did not pick up nan values as a string
```

```python
# Check to see if nan values are present in 'team'

race_times.isna().sum()
```

```python
# Cast all values in the 'team' to be a string

race_times['team'] = race_times['team'].astype(str)
```

```python
race_times['team'] = np.where(race_times['team']=='nan', pd.NA, race_times['team'])

race_times['team'].unique()
```

```python
race_times.head()
```

```python
# uci number is a unqiue identifier, does not need to have numeric operations performed on it 

race_times['uci_number'] = race_times['uci_number'].astype(str)
race_times['nationality'] = race_times['nationality'].astype(str)
```

```python
# change 'yob' to be a recognised year in date-time format
race_times['yob'] = race_times['yob'].astype(str)

race_times['yob'] = pd.to_datetime(race_times['yob'], format='%Y')
```

```python
race_times.dtypes
```

```python
# Change all split times into date time format

race_times['split_1'] = pd.to_datetime(race_times['split_1'], format='%M:%S.%f')
race_times['split_2'] = pd.to_datetime(race_times['split_2'], format='%M:%S.%f')
race_times['split_3'] = pd.to_datetime(race_times['split_3'], format='%M:%S.%f')
race_times['split_4'] = pd.to_datetime(race_times['split_4'], format='%M:%S.%f')
```

```python
race_times.head()
```

```python
# Change final time to date time format

race_times['time'] = pd.to_datetime(race_times['time'], format='%M:%S.%f')
```

```python
# Cast time off winner to a string then convert to date time format

race_times['tow'] = race_times['tow'].astype(str)
race_times['tow'] = pd.to_datetime(race_times['tow'], format='%S.%f')
```

```python
race_times.head()
```

```python
# Adding name identifiter of race and category

race_times['race_name'] = 'Lourdes Mens Elite'
```

```python
race_times['race_date'] = pd.to_datetime('2022-03-27')
```
